### **1. Fundamentos de bases de datos y para ciencia de datos.**

Una **base de datos (BD)** es una colección organizada de datos lógicamente relacionados, los cuales se almacenan de forma estandarizada. Para definir los datos se utilizan modelos y para su manipulación un lenguaje. Diferentes modelos se proponen buscando una mayor expresión para representar el mundo real. Las BD comprenden entidades, atributos y relaciones de la información.

Los lenguajes de manipulación de datos tienen como propósito ofrecer facilidad, simplicidad, y flexibilidad a la hora de utilizarlos para actualizar y recuperar información desde la base de datos.

Un modelo de datos está integrado por conjuntos de estructuras de datos, reglas u operadores para manipular los datos y un conjunto de reglas de integridad para asegurar estados consistentes.


### **2. Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

Un almacén de datos (AD) es un repositorio central de información. Los almacenamientos de datos son fundamentales como herramientas para extraer información, monitorear desempeños, analizar datos históricos, mejorar la calidad, coherencia y precisión de datos, tomar y respaldar decisiones fundamentales; permitiendo guardar los datos de manera eficaz y entregando resultados de consultas rápidamente a usuarios de manera simultánea.

El AD puede contener varias bases de datos, en cada BD los datos se organizan en tablas y columnas, dentro de cada columna se pueden definir una descripción de los datos. Las tablas se pueden organizar dentro de esquemas que se conciben como carpetas. Las herramientas de consulta usan el esquema para determinar a qué tablas de datos se obtiene acceso y en cuales de ellas se centrará el análisis.

**X1**: Monto del crédito otorgado (dólar NT): incluye tanto el crédito de consumo individual como su crédito familiar (complementario)
